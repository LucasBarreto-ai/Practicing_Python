```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dados = {'Nome': ['Lucas', 'Ginocleiton', 'Tanabares'],
         'Idade': [29, 36, 21],
         'Profissao': ['Banker', 'mecanico', 'LIXEIro']
        }
df = pd.DataFrame(dados)
```
